```python
gemini = os.getenv("gemini")
```

```python
from google import genai

client = genai.Client(api_key=gemini)

response = client.models.generate_content(
    model="gemini-2.0-flash", contents="Explain how AI works in a few words"
)
print(response.text)
```

```python
results = []
for idx, row in df.iterrows():
    prompt = f"""Extract the key topics, themes, or areas of focus from the following cybersecurity conference TITLE:

TITLE: "{row['Conference Name']}"

Return a list of key topics as list items, no explanations.
"""
    response = client.models.generate_content(
    model="gemini-2.0-flash", contents= prompt
)
    
    topics = response.text
    results.append(topics)
```

```python
df["Key Topics"] = results
```

```python
df["Key Topics"] = df["Key Topics"].str.replace(r"(?i)^here['’]s.*?:", "", regex=True).str.strip()
```
