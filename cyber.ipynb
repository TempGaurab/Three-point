```python
import pandas as pd
df = pd.read_csv("conference/cybersecurity_conferences.csv")
df
```

```python
import os
from dotenv import load_dotenv
```

```python
# Load environment variables
load_dotenv()
```

```python
gemini = os.getenv("gemini")
```

```python
from google import genai

client = genai.Client(api_key=gemini)

response = client.models.generate_content(
    model="gemini-2.0-flash", contents="Explain how AI works in a few words"
)
print(response.text)
```

```python
results = []
```

```python
df2 = df.head(14)
df2
```

```python
for idx, row in df2.iterrows():
    prompt = f"""Extract the key topics, themes, or areas of focus from the following cybersecurity conference description:

Description: "{row['Details']}"

Return a list of key topics as list items, no explanations.
"""
    response = client.models.generate_content(
    model="gemini-2.0-flash", contents= prompt
)
    
    topics = response.text
    results.append(topics)
```

```python
df2["Key Topics"] = results
```

```python
print(df2[["Title", "Key Topics"]])
```

```python
df3 = df.iloc[14:]
df3
```

```python
results = []
for idx, row in df3.iterrows():
    prompt = f"""Extract the key topics, themes, or areas of focus from the following cybersecurity conference description:

Description: "{row['Details']}"

Return a list of key topics as list items, no explanations.
"""
    response = client.models.generate_content(
    model="gemini-2.0-flash", contents= prompt
)
    
    topics = response.text
    results.append(topics)
```

```python
df3["Key Topics"] = results
```

```python
df_merged = pd.concat([df2, df3], ignore_index=True)
df_merged
```

```python
df_merged["Key Topics"] = df_merged["Key Topics"].str.replace(r"(?i)^here['’]s.*?:", "", regex=True).str.strip()
df_merged
```

```python
for point in df_merged['Key Topics']:
    print(point)
```

```python
df_merged = df_merged[['Title' , 'Link' , 'Key Topics']]
df_merged
```

```python
df_merged.to_csv("Cybersecutity-conferences-extracted.csv", index = False)
```
