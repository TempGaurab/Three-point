```python
import pandas as pd
import re
```

```python
df = pd.read_csv("merged_3.csv")
df
```

```python
# Clean and extract keywords
def extract_keywords(text):
    # Remove leading asterisks and newline characters
    keywords = re.findall(r'\*?\s*([\w\s&-]+)', text)
    return [kw.strip().lower() for kw in keywords if kw.strip() and "based on" not in kw.lower()]
```

```python
df['Keywords'] = df['Key Topics'].apply(extract_keywords)
```

```python
from collections import Counter
# Group by Type and find most common keywords
# Count keyword frequencies by category
grouped = df.groupby('Type')['Keywords'].sum()
keyword_counts = []
```

```python
for category, keywords in grouped.items():
    for keyword, count in Counter(keywords).items():
        keyword_counts.append({
            'Category': category,
            'Keyword': keyword,
            'Count': count
        })
```

```python
# Create and display the table
keyword_df = pd.DataFrame(keyword_counts).sort_values(by=['Category', 'Count'], ascending=[True, False])
keyword_df
```

```python
from tabulate import tabulate
```

```python
print(tabulate(keyword_df, headers='keys', tablefmt='fancy_grid'))
```
