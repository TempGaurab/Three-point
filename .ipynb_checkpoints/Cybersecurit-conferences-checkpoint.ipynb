```python
import requests as req
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup((req.get("https://www.securitymetrics.com/blog/top-cybersecurity-conferences-attend")).content, "html.parser")
soup
```

```python
# Step 2: Find all h3 tags and their content
data = []
all_h3s = soup.find_all("h3")
```

```python
for h3 in all_h3s:
    title_tag = h3.find("a")
    if title_tag:
        title = title_tag.text.strip()
        link = title_tag["href"]
        details = []

        # Traverse next siblings until the next h3 tag
        for sibling in h3.find_next_siblings():
            if sibling.name == "h3":
                break
            if sibling.name == "p":
                details.append(sibling.get_text(strip=True))

        data.append({
            "Title": title,
            "Link": link,
            "Details": " ".join(details)
        })
```

```python
import pandas as pd
df = pd.DataFrame(data)
df
```

```python
df.to_csv("cybersecurity_conferences.csv", index=False)
```
