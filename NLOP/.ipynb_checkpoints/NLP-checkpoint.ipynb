```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.patches import Patch

# --- Color scheme ---
colors = {
    'CS': '#FF8B94',             # CyberSecurity
    'AI-ML': '#A8E6CF',
    'HI': '#C7CEEA'               # Health Informatics
}

# --- Data (replace these with your actual sets) ---
data = {
    'CS': sorted(list(cyber)),        # CyberSecurity
    'AI-ML': sorted(list(ai_ml)),
    'HI': sorted(list(health))         # Health Informatics
}

# --- Identify intersections ---
cyber_set = set(data['CS'])
ai_set = set(data['AI-ML'])
health_set = set(data['HI'])

all_three = cyber_set & ai_set & health_set
cyber_ai = (cyber_set & ai_set) - all_three
cyber_health = (cyber_set & health_set) - all_three
ai_health = (ai_set & health_set) - all_three
cyber_only = cyber_set - ai_set - health_set
ai_only = ai_set - cyber_set - health_set
health_only = health_set - cyber_set - ai_set

# --- Create graph ---
G = nx.Graph()

# Add hubs
for hub in data.keys():
    G.add_node(hub, node_type='hub')

# Add subtopics
def add_subtopics(subtopics, intersection_type):
    for s in subtopics:
        G.add_node(s, node_type='subtopic', intersection_type=intersection_type)

add_subtopics(all_three, 'all_three')
add_subtopics(cyber_ai, 'cyber_ai')
add_subtopics(cyber_health, 'cyber_health')
add_subtopics(ai_health, 'ai_health')
add_subtopics(cyber_only, 'cyber_only')
add_subtopics(ai_only, 'ai_only')
add_subtopics(health_only, 'health_only')

# Connect hubs to subtopics
for hub, subtopics in data.items():
    for s in subtopics:
        G.add_edge(hub, s)

# --- Position nodes ---
pos = {}

# Fixed hub positions (triangle layout)
hub_pos = {
    'AI-ML': np.array([-2.5, 0.8]),   
    'CS': np.array([2.5, 0.8]),       
    'HI': np.array([0, -2.8])         
}

pos.update(hub_pos)

# --- Intersections positioning ---
center = np.array([0, -0.5])

# --- All-three intersection nodes (center cluster) ---
n_all_three = len(all_three)
all_three_center = center + np.array([0, 0])  

if n_all_three == 1:
    pos[list(all_three)[0]] = all_three_center
elif n_all_three > 1:
    angles = np.linspace(0, 2 * np.pi, n_all_three, endpoint=False)
    radius = 0.35
    for i, node in enumerate(all_three):
        pos[node] = all_three_center + radius * np.array([np.cos(angles[i]), np.sin(angles[i])])


# --- Pairwise intersections (much larger offsets to prevent overlap) ---
# Position them farther away and in clear sectors
intersection_map = {
    'cyber_ai': np.array([0.0, 1.5]),      # moved way up
    'cyber_health': np.array([1.8, -1.2]), # moved further bottom-right
    'ai_health': np.array([-1.8, -1.2])    # moved further bottom-left
}

for inter_type, offset in intersection_map.items():
    nodes = list(eval(inter_type))
    n_nodes = len(nodes)
    if n_nodes == 0:
        continue
    cluster_center = center + offset
    angles = np.linspace(0, 2*np.pi, n_nodes, endpoint=False)
    radius = 0.75  # increased for more spacing
    for i, node in enumerate(nodes):
        pos[node] = cluster_center + radius * np.array([np.cos(angles[i]), np.sin(angles[i])])

# --- Unique nodes around hubs ---
unique_map = {
    'CS': (list(cyber_only), (1, 0)),     
    'AI-ML': (list(ai_only), (-1, 0)),    
    'HI': (list(health_only), (0, -1))    
}

for hub, (nodes, direction) in unique_map.items():
    hub_center = hub_pos[hub]
    for n in nodes:
        base_angle = np.arctan2(direction[1], direction[0])
        angle = np.random.uniform(base_angle - np.pi/5, base_angle + np.pi/5)
        r = np.random.uniform(1.5, 1.75)
        pos[n] = hub_center + r * np.array([np.cos(angle), np.sin(angle)])


# --- Draw ---
fig, ax = plt.subplots(figsize=(14, 10), facecolor='#fafafa')

# Edges
nx.draw_networkx_edges(G, pos, ax=ax, alpha=0.12, edge_color='gray')

# Hubs
hub_nodes = list(hub_pos.keys())
nx.draw_networkx_nodes(
    G, pos, nodelist=hub_nodes,
    node_color=[colors[h] for h in hub_nodes],
    node_size=2600, edgecolors='black', linewidths=2.5
)
nx.draw_networkx_labels(G, pos, labels={h: h for h in hub_nodes},
                        font_size=13, font_weight='bold')

# Subtopics
node_colors = {
    'all_three': '#FF6B6B',
    'cyber_ai': '#4ECDC4',
    'cyber_health': '#95E1D3',
    'ai_health': '#FFE66D',
    'cyber_only': colors['CS'],
    'ai_only': colors['AI-ML'],
    'health_only': colors['HI']
}
sub_nodes = [n for n, d in G.nodes(data=True) if d['node_type'] == 'subtopic']
sub_colors = [node_colors[G.nodes[n]['intersection_type']] for n in sub_nodes]

nx.draw_networkx_nodes(
    G, pos, nodelist=sub_nodes,
    node_color=sub_colors, node_size=340,
    edgecolors='black', linewidths=0.8, alpha=0.9
)
nx.draw_networkx_labels(G, pos, labels={n: n for n in sub_nodes},
                        font_size=7, font_weight='medium')

# --- Final touches ---
ax.set_title("AI, CS, and HI", fontsize=15, fontweight='bold')
ax.axis('off')
plt.tight_layout()
plt.show()
```
