```python
import pandas as pd
import numpy as np
import requests
```

```python
url = 'https://aideadlin.es/?sub=ML,CV,CG,NLP,RO,SP,DM,AP,KR,HCI'
response = requests.get(url)
response
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(response.content, "html.parser")
soup
```

```python
conference_divs = soup.find_all("div", class_="ConfItem")
conference_divs
```

```python
conference_data = []
```

```python
for conf in conference_divs:
    # Extract core fields
    title = conf.select_one(".conf-title a")
    title_small = conf.select_one(".conf-title-small a")
    url = conf.select_one(".conf-title-icon a[href]")
    date = conf.select_one(".conf-date")
    location = conf.select_one(".conf-place a")
    note_div = conf.select_one(".note")
    
    conference_data.append({
        "Title": title.text.strip() if title else "",
        "Short Name": title_small.text.strip() if title_small else "",
        "Conference URL": url['href'] if url else "",
        "Date": date.text.strip() if date else "",
        "Location": location.text.strip() if location else "",
        "Note": note_div.get_text(strip=True) if note_div else ""
    })
```

```python
df = pd.DataFrame(conference_data)
df
```

```python
df.to_csv("extracted_AI_ML_conferences.csv", index=False)
```
